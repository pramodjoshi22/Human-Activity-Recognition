```python
# Get data
from google.colab import drive
drive.mount('/content/drive')
%cd '/content/drive/My Drive/CS3244 Team37'

x_train = pd.read_csv('./HAPT Data Set/x_train_upsampled.csv')
y_train = pd.read_csv('./HAPT Data Set/y_train_upsampled.csv')
# x_train = pd.read_csv('./HAPT Data Set/x_train.csv')
# y_train = pd.read_csv('./HAPT Data Set/y_train.csv')
x_test = pd.read_csv('./HAPT Data Set/x_test.csv')
y_test = pd.read_csv('./HAPT Data Set/y_test.csv')
```

```python
# Preprocess data
pipe = Pipeline([
    # 'preproceesors'
    ('model', neighbors.KNeighborsClassifier(n_neighbors = 6))
])

parameters = {
    "model__n_neighbors" : range(1, 20, 1)
}

# TODO: figure out how to get make_scorer to work with pipeline (default is accuracy)
# auc_roc_scorer = make_scorer(roc_auc_score)
# grid_search = GridSearchCV(pipe, parameters, n_jobs=-1, verbose=1, scoring = auc_roc_scorer)
grid_search = GridSearchCV(pipe, parameters, n_jobs=-1, verbose=1)

grid_search.fit(x_train, y_train)
print(grid_search.best_params_)
print(grid_search.best_score_)
grid_search.score(x_test, y_test)
```
